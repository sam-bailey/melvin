```python
import jax.numpy as jnp
from melvin.minimization_problem import MinimizationProblem
from jax.experimental.optimizers import adam, sgd
import jax
import matplotlib.pylab as plt
from functools import partial
```

```python
x0 = jnp.array([5.0, -2.0])

def objective_fn(x, data):
    return x[0]**2 + (x[1] - 1.0)**2

model = MinimizationProblem(
    "Quadratic Function",
    objective_fn, 
    initial_params=x0,
    optimizer=sgd
)
model.fit(n_steps=200, data=None)
```

```python
ax = model.plot_history()
plt.yscale("log")
plt.show()
```
